```python
# importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
#sending get request to the webpage server to get the source code of the page
page = requests.get("https://presidentofindia.nic.in/former-presidents.htm")
```

```python
page
```

```python
#page content
soup = BeautifulSoup(page.content)
soup
```

```python
#scrapping first movie name
former_president = soup.find ("div", class_= "presidentListing")
former_president
```

```python
former_president.text

A= former_president.text
A
A.replace ('\n', " ")
```

```python
# scrapping multiple former president details
Former_president_details = []

for i in soup.find_all ("div", class_= "presidentListing"):
    Former_president_details.append(i.text)

Former_president_details
```

```python
#Making dataframe
import pandas as pd
df = pd.DataFrame ({'Former Preseident Details': Former_president_details})
df
```
